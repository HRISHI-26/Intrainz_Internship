#Credit Card Fraud Detection

Importing Libraries

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
#Loading Dataset
data = pd.read_csv("creditcard.csv")
```

```python
#first five rows of the dataset
data.head()
```

```python
#last 5 rows
data.tail()
```

```python
#information about the dataset
data.info()
```

```python
#Checking the number of missing values in each column

data.isnull().sum()
```

```python
#distribution of legit transcations and fraudulent transactions
data['Class'].value_counts()
```

This dataset is highly unbalanced

```python
#seperating data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

As per above result we have high variations so we are Creating new dataset with uniformity

```python
#compare the values for both transactions
data.groupby('Class').mean()
```

Under-Sampling


Build a sample dataset ccontaining similar distribution of normal transactions and fraudulent transactions

```python
#taking random 492 transactions to balance with fraudulent in sample dataset we create
legit_sample = legit.sample(n=492)
```

```python
#concatenate two DataFrames to create new dataset

new_dataset = pd.concat([legit_sample, fraud], axis=0) #axis = 0 --> row
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the Data into features and targets

```python
X = new_dataset.drop(columns = 'Class', axis=1) #axis == 1 --> column
Y = new_dataset['Class']
```

```python
X #No class column
```

Split into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X_train, X_train.shape, X_test.shape)
```

Model Training - Logistic Regression

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Accuracy score
X_train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(train_accuracy)
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_accuracy)
```

If the train_accuracy & test_accuracy is almost equal then its a good model
